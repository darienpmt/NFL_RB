# Finding optimal lambda for Ridge and LASSO

```python
# target and features
X = smaller_df.drop('NFL Y/G', axis=1)
y = smaller_df['NFL Y/G']

# splitting the data into train/test
X, X_test, y, y_test = train_test_split(X, y, test_size=.2, random_state=60)

X = np.array(X)
y = np.array(y)

# will divide data into 5 train/test sets
kf = KFold(n_splits=5, shuffle=True, random_state = 71)

# the possible alphas we will consider
alphas = 10**np.linspace(-3,3,500)
```

```python
# collect mean absolute errors
lasso_av_errors = []
ridge_av_errors = []

# collect r_2s
lasso_r2_avs = []
ridge_r2_avs = []

# loop through the possible alphas
for curr_alpha in alphas:
    
    lasso_r2s = []
    ridge_r2s = []
    
    lasso_errors = []
    ridge_errors = []
    
    # loop through different folds
    for train_ind, val_ind in kf.split(X,y):
        
        # separates the training and validation data
        X_train, y_train = X[train_ind], y[train_ind]
        X_val, y_val = X[val_ind], y[val_ind] 
        
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_val_scaled = scaler.transform(X_val)
        
        # creates the LASSO model "object"
        lasso_model = Lasso(alpha = curr_alpha, max_iter=5000, tol=40)
        ridge_model = Ridge(alpha = curr_alpha, max_iter=5000, tol=40)
        
        # fit the training data to a LASSO model
        lasso_model.fit(X_train_scaled, y_train)
        ridge_model.fit(X_train_scaled, y_train)
        
        # add the mae to the list of errors
        lasso_errors.append(mae(y_val, lasso_model.predict(X_val_scaled)))
        ridge_errors.append(mae(y_val, ridge_model.predict(X_val_scaled)))
        
        # add the r_2 to the list of r2s
        lasso_r2s.append(lasso_model.score(X_val_scaled, y_val))
        ridge_r2s.append(ridge_model.score(X_val_scaled, y_val))
    
    # append the average r2 from our cv for each alpha
    lasso_r2_avs.append(np.mean(lasso_r2s))
    ridge_r2_avs.append(np.mean(ridge_r2s))
    
    # append the average mae from our cv for each alpha
    lasso_av_errors.append(np.mean(lasso_errors))
    ridge_av_errors.append(np.mean(ridge_errors))

# finds the best average r2
lasso_best_r = max(lasso_r2_avs)
ridge_best_r = max(ridge_r2_avs)

#finds the best mae
lasso_best_mae = min(lasso_av_errors)
ridge_best_mae = min(ridge_av_errors)

# finds the alpha with the best r2
lasso_best_alpha_r = alphas[lasso_r2_avs.index(lasso_best_r)]
ridge_best_alpha_r = alphas[ridge_r2_avs.index(ridge_best_r)]

# finds the alpha with the best mae
lasso_best_alpha_mae = alphas[lasso_av_errors.index(lasso_best_mae)]
ridge_best_alpha_mae = alphas[ridge_av_errors.index(ridge_best_mae)]
```
